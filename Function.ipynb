```python
def prt_cone_vol(r, h) :
    if r > 0 and h > 0 :
        # r, h 모두 양수일 때
        vol = 1/3 * 3.14 * r ** 2 * h
        print("원뿔의 부피는", vol, "입니다.")
    else :
        # r, h가 음수일 때
        print("반지름과 높이 값에 양수를 입력하세요.")
```

## 원뿔 부피 계산 함수 호출

```python
radius = 20
height = 30

prt_cone_vol(radius, height)
```
