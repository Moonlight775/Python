```python
def prt_cone_vol(r, h) :
    if r > 0 and h > 0 :
        # r, h 모두 양수일 때
        vol = 1/3 * 3.14 * r ** 2 * h
        print("원뿔의 부피는", vol, "입니다.")
    else :
        # r, h가 음수일 때
        print("반지름과 높이 값에 양수를 입력하세요.")
```

## 원뿔 부피 계산 함수 호출

```python
radius = 20
height = 30

prt_cone_vol(radius, height)
```

## 숫자를 입력받고 역순으로 출력하는 함수를 사용한 프로그램

```python
def reverse_number(num) :
    while num != 0 :
        digit = num % 10
        num = num // 10
        print(digit, end = "")

num = int(input("정수를 입력해주세요 : "))
reverse_number(num)
```
