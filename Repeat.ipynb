## 사용자가 입력한 수 까지의 합 출력

```python
num = int(input("어디까지 더할까요? "))

sum = 0
i = 1

while i <= num :
    sum = sum + i
    i = i + 1

print(sum)
```

## 사용자가 입력한 단의 구구단 출력

```python
dan = int(input("출력할 단을 입력하세요: "))

i = 1

while i <= 9:
    print('%d X %d = %d' %(dan, i, dan*i))
    i = i + 1
```

## 반지름과 높이가 (10, 1), (20, 5), (30, 14)인 원뿔의 부피와 겉넓이를 각각 출력

```python
rad_list = range(10, 31, 10)
hei_list = [1, 5, 14]

for rad, hei in zip(rad_list, hei_list):        # 여러 리스트의 값들을 하나씩 엮어서 가져오는 zip 함수
    # 부피 & 겉넓이 계산
    vol = 1/3 * 3.14 * rad ** 2 * hei
    surf = 3.14 * rad ** 2 + 3.14 * rad * hei
    print("반지름", rad, "높이", hei, "원뿔")
    print("원뿔의 부피는", vol, "입니다.")
    print("원뿔의 겉넓이는", surf, "입니다.")
```

## 구구단 표를 출력하는 문제

```python
print(format("구구단표", ">20s"))

# 표 머리 출력
print("   |", end = "")
for j in range(1, 10):
    print("   ", j, end = "")

# 새로운 행 삽입
print()
print("---------------------------------------------------")

# 구구단 표 출력

for i in range(1, 10):
    print(i, " |", end ="")
    for j in range(1, 10):
        print(format(i*j, ">5d"), end ="")
    print()
```
