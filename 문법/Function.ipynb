## 숫자를 입력받고 역순으로 출력하는 함수를 사용한 프로그램

```python
def reverse_number(num) :
    while num != 0 :
        digit = num % 10
        num = num // 10
        print(digit, end = "")

num = int(input("정수를 입력해주세요 : "))
reverse_number(num)
```

## 원뿔 부피 계산 결과 반환 함수 정의

```python
def rtn_cone_vol(r, h) :
    if r > 0 and h > 0 :
        # r, h 모두 양수일 때
        vol = 1/3 * 3.14 * r ** 2 * h
        return vol
    else :
        # r, h가 음수일 때
        print("반지름과 높이 값에 양수를 입력하세요.")

print(format(rtn_cone_vol(10, 20), "<10.3f"), "입니다.")
```

## 원뿔 부피 및 겉넓이 반환 함수 정의

```python
def rtn_cone_vol_surf(r, h) :
    if r > 0 and h > 0 :
        # r, h 모두 양수일 때
        vol = 1/3 * 3.14 * r ** 2 * h
        surf = 3.14 * r ** 2 + 3.14 * r * h
        return vol, surf
    else :
        # r, h가 음수일 때
        print("반지름과 높이 값에 양수를 입력하세요.")

vol1, surf1 = rtn_cone_vol_surf(50, 100)
print("부피는", vol1, "입니다.")
print("겉넓이는", surf1, "입니다.")
```
